```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
```

```python
data=pd.read_csv('/content/diabetes.csv')
```

```python
x=data.drop(columns='Outcome')
y=data['Outcome']
```

```python
scalar=StandardScaler()
```

```python
scalar.fit(x)
```

```python
standardized_data=scalar.transform(x)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
scalar=StandardScaler()
```

```python
classifier=svm.SVC(kernel='linear')
```

```python
classifier.fit(x_train,y_train)
```

```python
x_train_pr = classifier.predict(x_train)
training_data_accuracy=accuracy_score(y_train,x_train_pr)
```

```python
print(f"the training data  accuracy is {training_data_accuracy}")
```

```python
x_test_pr = classifier.predict(x_test)
test_data_accuracy=accuracy_score(y_test,x_test_pr)
```

```python
print(f"the testing  data  accuracy is {test_data_accuracy}")
```
